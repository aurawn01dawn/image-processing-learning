### matplotlib

```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
```

```python
img = mpimg.imread("images/flower_cat.jpg")
```

```python
# default numpy array
print(type(img))
```

```python
print(img.shape)
```

```python
plt.imshow(img)
plt.colorbar()
```

### skimage （scikit-image）

```python
from skimage import io
```

```python
image = io.imread("images/cats.jpg")
print(type(image))
print(image.shape)
```

```python
plt.imshow(image)
# plt.colorbar()
```

```python
print(image)
```

```python
from skimage import img_as_float, img_as_ubyte
```

```python
#
image_float = img_as_float(image)
print(image_float)
```

### opencv

```python
import cv2
```

```python
# color changed bgr not rgb
grey_image = cv2.imread("images/dogs.jpg")
plt.imshow(grey_image)
```

```python
# color changed
grey_image2 = cv2.imread("images/dogs.jpg", 0)
plt.imshow(grey_image2)
```

```python
# need some time
cv2.imshow("Grey Image", grey_image)
color_image = cv2.imread("images/dogs.jpg")
cv2.imshow("Color Image", color_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
